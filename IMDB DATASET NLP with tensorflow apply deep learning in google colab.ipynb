```
import tensorflow_datasets as tfds # add tensorflow inbuild datset
imdb, info = tfds.load("imdb_reviews", with_info=True, as_supervised=True) # load method for load data set into imdb, info
#info = meta data
# imbd =main dataset.
# its by default dataset so some numpay operations required of this data .becoz we dont know how this data will precced becoz it by defualt dataset of tensorflow
print(imdb)
```

```
import numpy as np # import numpy to deal linear algebra
print(imdb)# its show meta data
print(info)# its show some info not dataset
train_data, test_data = imdb['train'], imdb['test'] # spilte 25000 record for training and 25000 for testing  

training_sentences = [] # its for senetences
training_labels = [] # its for label output for particular sentences

testing_sentences = [] # same
testing_labels = [] # same

# str(s.tonumpy()) is needed in Python3 instead of just s.numpy()
for s,l in train_data: # here this loop goes till 25000
  training_sentences.append(s.numpy().decode('utf8')) # here you use numpy array becoz its inbuild data , so inka alag process hai extracte krnai ka dont go in depth.  
  training_labels.append(l.numpy()) # extract labels
  
for s,l in test_data: # its for test loop will go till 25000 test data
  testing_sentences.append(s.numpy().decode('utf8')) # same 
  testing_labels.append(l.numpy()) # same
  
training_labels_final = np.array(training_labels) # training labels converts into numpy homogenous single dimesional array or vrctor....
testing_labels_final = np.array(testing_labels) # same but its testing 
```

```
# after extract data from inbuild data set of tensorflow and time to preprocessing ,padding ,tokenize
# tokenize mean words convert into some  number / mean assign some no to the words that is called tokenaziations. 
vocab_size = 10000
embedding_dim = 16
max_length = 120
trunc_type='post'
oov_tok = "<OOV>"

# we knew all about it.
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(training_sentences)
padded = pad_sequences(sequences,maxlen=max_length, truncating=trunc_type)

testing_sequences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(testing_sequences,maxlen=max_length)
```

```
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

def decode_review(text):
    return ' '.join([reverse_word_index.get(i, '?') for i in text])

print(decode_review(padded[3]))
print(training_sentences[3])
```

```
# now perform ANN using keras and tensorflow
# we all knew about it in ANN practical you can go for it on github and deep learning , how ANN implements.
# you can implement ANN like this or in another way using add keywords go and take a look of my implementation from ANN for second way.
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length), # its first emedding layer, this is another type of keras layer , this layer use in NLP
    
    tf.keras.layers.Flatten(),# its flatten which communicate between simple layer to dense layer and it convert N dimesional array into single D vector , becoz  dense 1 D vector take as input.
    
    tf.keras.layers.Dense(6, activation='relu'), # its fully connected layer with 6 nureons and relu actiivation fun.
    tf.keras.layers.Dense(1, activation='sigmoid') # its output layer with single nuerons using sigmoid function.
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) # here compile this declare and made ANN using loss function and optimizer and metrics
# we knew go and take a look ANN implements on github ->deep leaning 
model.summary() # its all about summery of ANN , look at deeply.
```

```
num_epochs = 10
model.fit(padded, training_labels_final, epochs=num_epochs, validation_data=(testing_padded, testing_labels_final))
```
